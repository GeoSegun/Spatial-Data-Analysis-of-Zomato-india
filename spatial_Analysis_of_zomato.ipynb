## Extracting latitude and longitude values

```python
!pip install geopy
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent='app')
```

```python
lat=[]
lon=[]
for location in locations['Name']:
    location= geolocator.geocode(location)
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
#adding the extracted location Values to the dataframe created
locations['latitude']= lat
locations['longitude']= lon
```

```python
locations.head(10)
```

```python
rest_restuarant = zdata['location'].value_counts().reset_index()
rest_restuarant
```

```python
rest_restuarant.columns = ['Name', 'count']
rest_restuarant.head()
```

```python
resturant_location = rest_restuarant.merge(locations, on='Name', how= 'left').dropna()
resturant_location.head()
```

## Marker cluster Analysis of the resturant Rating and the HeatMap

```python
zdata['rate'].unique() 
```

```python
zdata.dropna(axis=0, subset= ['rate'], inplace=True)
```

```python
#function the split values
def split(x):
    return x.split('/')[0]
```

```python
zdata['rating'] = zdata['rate'].apply(split)
```

```python
zdata.head()
```

```python
zdata['rating'].unique()
```

```python
zdata.replace('NEW', 0, inplace= True)
zdata.replace('-', 0, inplace= True)
```

```python
zdata.head()
```

```python
zdata.dtypes #checking the various data types
```

```python
zdata['rating']= pd.to_numeric(zdata['rating'])
```

```python
avg_rating = zdata.groupby('location')['rating'].mean().sort_values(ascending= False).values
```

```python
location = zdata.groupby('location')['rating'].mean().sort_values(ascending= False).index
```

```python
len(avg_rating)
```

```python
len(lon)
```

```python
rating = df()
```

```python
lat=[]
lon=[]
for loc in location:
    loc= geolocator.geocode(loc)
    if loc is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(loc.latitude)
        lon.append(loc.longitude)
```

```python
rating['location'] = location
rating['lat'] = lat
rating['lon'] = lon
rating['Average rating'] = avg_rating

rating.head()
```

```python
rating.dropna(inplace= True)
```

```python
HeatMap(rating[['lat', 'lon', 'Average rating']], zoom= 10).add_to(basemap)
```

```python
basemap
```

From the Map above, Resturant at the central path of India has more Customer rating compared to other reason. the rating increases as you go Northward.

```python
#this function automate the various cuisines present in the zomato data

def heatmap_zone(zone):
    filter= zdata['cuisines']== zone
    zdata3=zdata[filter]
    zdata_zone= zdata3.groupby('location')['url'].count().reset_index().dropna()
    zdata_zone.columns= ['Name', 'count']
    zdata_zone= zdata_zone.merge(locations, on= 'Name', how= 'left')
    HeatMap(zdata_zone[['latitude', 'longitude', 'count']], zoom= 10).add_to(basemap)
    return basemap
    
```
